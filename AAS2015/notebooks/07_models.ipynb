# Basic Models and Fitting

**Goals:**

- Decouple code for models, fitting statistics, and optimizers
- Make it easier to add models that take full advantage of fitting machinery without knowledge of fitting engine details
- Allow arbitrary combinations of models to produce more complex models
- Support basic parameter constraints
- Don't try to solve all possible problems

```python
from __future__ import print_function, division
from astropy.modeling import models, fitting
import numpy as np
```

```python
# this will make plots appear inside the notebbok
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
dir(models)
```

## Simple one component model

### Create basic 1D Gaussian model

```python
g = models.Gaussian1D(amplitude=1.1,mean=0.0,stddev=0.2)
print(g.amplitude, g.mean, g.stddev)
```

Models evaluate just like functions

```python
x = np.linspace(-1.,1.,500) # array of equally spaced x  values from -1 to 1
plt.plot(x,g(x))
None
```

### Modifying model parameters

```python
g.mean = 0.5
print(g.stddev)
plt.plot(x,g(x))
None
```

## Simple Fitting Example

Create data to fit to

```python
data = g(x) + np.random.normal(0., 0.4, x.shape)
plt.plot(x,data)
None
```

```python
fitter = fitting.LevMarLSQFitter()
# change initial model values
g.mean = .2 
g.amplitude = 2
g.stddev = 0.05
fit_g = fitter(g, x, data)
print(fit_g)
```

```python
plt.plot(x,fit_g(x))
None
```

## Compound Models

*Examples involving compound models require astropy v1.0, due out within a month*

```python
g1 = models.Gaussian1D(1,1,.1)
g2 = models.Gaussian1D(.5,.8,.2)
g = g1 + g2
```

```python
plt.clf()
x = np.linspace(0.,2,2000)
y = g(x) + np.random.normal(0,.2,x.shape)
plt.plot(x,y)
None
```

```python
g3 = models.Gaussian1D(1,1,.1)
g4 = models.Gaussian1D(1,1,.1)
gf = g3+g4
```

```python
fitter = fitting.SLSQPLSQFitter()
```

```python
fitmodel = fitter(gf, x, y)
```

```python
plt.plot(x,g(x),color='y')
plt.plot(x,fitmodel(x),color='r')
None
```

## Parameter Constraints

## Tied Parameters

```python
g = models.Gaussian1D(amplitude=1.1,mean=0.0,stddev=0.2)
x = np.linspace(-1.,1.,500) # array of equally spaced x  values from -1 to 1
gorig = g(x)
y = g(x) + np.random.normal(0,.2,x.shape)
fitter = fitting.LevMarLSQFitter()
g.amplitude = .9
g.amplitude.fixed = True
fit_g = fitter(g,x,y)

plt.plot(x,y)
plt.plot(x,gorig)
plt.plot(x,fit_g(x))
None
```

## Parameter Bounds

```python
g.amplitude.fixed = False
g.amplitude.max = .6
fit_g = fitter(g,x,y)

plt.plot(x,y)
plt.plot(x,gorig)
plt.plot(x,fit_g(x))
None
```

```python
# force ratio of means to be specified constant
def tied(model):
    mean_1 = 2. * model.mean_0
    return mean_1
g1 = models.Gaussian1D(1,1,.1)
g2 = models.Gaussian1D(.5,.6,.2)
g = g1 + g2
y = g(x) + np.random.normal(0.,.2,x.shape)
gf = models.Gaussian1D(.4,.6,.3) + models.Gaussian1D(1,1,.1)
gf.mean_1.tied = tied
fitter2 = fitting.SimplexLSQFitter()
fitmodel2 = fitter2(gf,x,y)
```

```python
plt.plot(x, y)
plt.plot(x, g(x), color='y')
plt.plot(x, fitmodel2(x), color='orange')
fitmodel2
```

## Supporting Ensemble Fits

```python
p = models.Polynomial1D(2, c0=.5, c1=0, c2=1) # one model
parr = models.Polynomial1D(2, c0=0.5*np.ones(10), c1=np.zeros(10), c2=np.ones(10), n_models=10)
parr
```

```python
parreal = models.Polynomial1D(2, c0=0.5*np.ones(10)+np.random.normal(0,0.1,10), \
                               c1=np.zeros(10), c2=np.ones(10)+np.random.normal(0,.1,10),
                               n_models=10)
x = np.linspace(-1,1,21)
data = parreal(x, model_set_axis=False)
data
modc = parreal.copy()
fitter = fitting.LinearLSQFitter()
fitmodels = fitter(parreal, x, data)
print(parreal)
print(fitmodels)
```
