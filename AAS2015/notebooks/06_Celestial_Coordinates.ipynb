# Celestial Coordinates

## Documentation

For more information about the features presented below, you can read the
[astropy.coordinates](http://docs.astropy.org/en/stable/coordinates/index.html) docs.

## About

Astropy includes a framework to represent celestial coordinates and transform
between them. Astropy 0.4 only includes a few common coordinate frames (ICRS,
FK4, FK5, and Galactic), but the soon-to-be-released v1.0 will include more built-in
coordinate frames (particularly AltAz coordinates).  Users are also free to define
custom coordinate systems that make sense for a particular domain 

## Creating coordinate objects

Creating coordinate objects is straightforward.  For typical use cases, you work with the `astropy.coordinates.SkyCoord` class.

```python
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
SkyCoord(ra=10.68458 * u.deg, dec=41.26917 * u.deg, frame='icrs')
```

In addition to passing angular Quantity objects, the SkyCoord class can parse coordinate strings: 

```python
SkyCoord('00h42m44.3s +41d16m9s', frame='icrs')
```

## Accessing angles

The individual components of a coordinate are ``Angle`` objects, and their
values are accessed using special attributes:

```python
c = SkyCoord('00h42m44.3s +41d16m9s', frame='icrs')
```

```python
c.ra
```

```python
c.ra.hour
```

```python
c.ra.hms
```

```python
c.dec
```

```python
c.dec.radian
```

These components are sub-classes of `Quantity` objects, so they can also be used the same way as a `Quantity`:

```python
c2 = SkyCoord('00h42m44.2s +41d16m9s', frame='icrs')
dang = (c.ra - c2.ra)/u.day
dang.to(u.arcmin/u.yr)
```

## Converting to other systems

To convert to some other coordinate frame (loosly but imprecisely sometimes called a "coordinate system"), the easiest method is to use
attribute-style access with short names for the built-in frames:

```python
c.galactic
```

```python
c.fk5
```

While these conveniences allow fast conversion to common coordinate systems, they do not allow any customization of the parameters of the new frame, such as the equinox. The more general way of transforming coordinates is via the ``transform_to`` method. In addition to taking convenience strings:

```python
c.transform_to('galactic')
```

this method can also take frame objects that can be customized:

```python
from astropy.coordinates import FK5

frame = FK5(equinox='J2010')
c.transform_to(frame)
```

## Name resolution

The ``astropy.coordinates`` subpackage also provides a quick way to get coordinates for
named objects (with an internet connection). All coordinate classes have a
special class method, `from_name()`, that accepts a string and queries [Sesame](http://cds.u-strasbg.fr/cgi-bin/Sesame) to retrieve coordinates for that
object:

```python
SkyCoord.from_name("M16")
```

## Using arrays in coordinates

Numpy arrays or lists can be used inside coordinate objects instead of scalar floating point values. This is much more efficient that creating a coordinate object for each source. 

```python
ra = [1, 2, 3, 4] * u.hourangle
dec = [5, 6, 7, 8] * u.deg
c = SkyCoord(ra, dec, frame='icrs')
c
```

As a less contrived case, the following example demonstrates how to combine the ``Table`` class with coordinate objects (you can download the data from [here](data/2mass.tbl)).

```python
from astropy.table import Table
```

```python
t = Table.read('data/2mass.tbl', format='ascii.ipac')
t
```

```python
c = SkyCoord(t['ra'] * u.deg, t['dec'] * u.deg, frame='fk5')
```

```python
c.ra.degree
```

```python
c = SkyCoord(t['clon'], t['clat'], frame='fk5')
c
```

## Doing Celestial to Alt-Az conversions

Astropy v1.0 adds the ability to determine where an object will be found on the sky as viewed from Earth, given a particular location and time.  First we define a location and a time at which we want to observe (for this first example, that's the local midnight).

```python
from astropy.time import Time
from astropy.coordinates import EarthLocation

# Bear Mountain State Park in New York, USA
bear_mountain = EarthLocation(lat=41.3*u.deg, lon=-74*u.deg, height=390*u.m)
utcoffset = -4*u.hour  # Eastern Daylight Time
midnight = Time('2012-7-13 00:00:00') - utcoffset
```

Now we need a celestial object to observe.  Let's try M33.

```python
m33 = SkyCoord.from_name('M33')
```

Then we just convert to an `AltAz` coordinate at that location and time

```python
from astropy.coordinates import AltAz

m33.transform_to(AltAz(obstime=midnight, location=bear_mountain))
```

We can also ask what M33's airmass will be over the course of the night:

```python
# this first line sets up the ipython notebook to make matplotlib plots inside the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#we also supress a warning about vector transforms - not relevant because we know this transform works
import warnings
warnings.filterwarnings('ignore',module='astropy.coordinates.baseframe')

delta_midnight = np.linspace(-2, 7, 100)*u.hour
m33altazs = m33.transform_to(AltAz(obstime=midnight+delta_midnight, location=bear_mountain))

plt.figure(figsize=(12,8))
plt.plot(delta_midnight, m33altazs.secz)
plt.xlim(-2, 7)
plt.ylim(1, 4)
plt.xlabel('Hours from EDT Midnight')
plt.ylabel('Airmass [Sec(z)]')
None
```

To go one more step, we can use the position of the sun over the night to determine when twilight occurs:

```python
from astropy.coordinates import get_sun

delta_midnight = np.linspace(-12, 12, 1000)*u.hour
altazframe = AltAz(obstime=midnight + delta_midnight, location=bear_mountain)
sunaltazs = get_sun(midnight + delta_midnight).transform_to(altazframe)
m33altazs = m33.transform_to(altazframe)

plt.figure(figsize=(12,8))
plt.plot(delta_midnight, sunaltazs.alt, color='y', label='Sun')
plt.scatter(delta_midnight, m33altazs.alt, c=m33altazs.az, label='M33', lw=0, s=8)
plt.fill_between(delta_midnight, 0, 90, sunaltazs.alt < -0*u.deg, color='0.5', zorder=0)
plt.fill_between(delta_midnight, 0, 90, sunaltazs.alt < -18*u.deg, color='k', zorder=0)
plt.colorbar().set_label('Azimuth [deg]')
plt.legend(loc='upper left')
plt.xlim(-12, 12)
plt.xticks(np.arange(13)*2 -12)
plt.ylim(0, 90)
plt.xlabel('Hours from EDT Midnight')
plt.ylabel('Altitude [deg]')
None
```
